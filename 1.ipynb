##### Библиотеки

```python
import numpy as np
import matplotlib.pyplot as plt
```

#### Генерируем исходный набор точек и целевые значения

```python
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)
```

#### Создаем модель линейной регрессии

```python
w = np.random.randn(1)
b = np.random.randn(1)
learning_rate = 0.1
n_iterations = 1000
m = len(X)
```

#### Обучаем модель линейной регрессии

В качестве смещения - производная MSE по весам

```python
for i in range(n_iterations):
    y_pred = w * X + b
    error = y_pred - y

    dw = (2/m) * np.sum(error * X)
    db = (2/m) * np.sum(error)

    w -= learning_rate * dw
    b -= learning_rate * db
```

#### Выводим результаты

```python
print(f"Обученные параметры: w = {w[0]:.3f}, b = {b[0]:.3f}")
plt.scatter(X, y, color="blue", label="Данные")
plt.plot(X, w * X + b, color="red", label="Линия регрессии")
plt.legend()
plt.xlabel("X")
plt.ylabel("y")
plt.title("Линейная регрессия")
plt.show()
```
